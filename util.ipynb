# Oein의 유틸 함수

아래의 코드를 자신의 주피터 파일에서 실행하여 함수를 정의후 사용하세요.

## inp(type, name?)

특정 타입만 입력받습니다.

```py
# 예제 사용
inp("int")
inp("str")
inp("float", "점수")
inp("bool", "성공여부")
```

### inp의 타입
 - int (정수)
 - str (문자열)
 - bool (참거짓)
 - float (소수)
 - char (문자)
 
 ## isIntStr(str)
 
 들어온 문자열이 정수인지 확인합니다.
 
 ```py
 isIntStr("1") # True
 isIntStr("123") # True
 isIntStr("123.123") # False
 isIntStr("HI") # False
 ```
 
 ## isFloatStr(str)
 
 들어온 문자열이 소수 혹은 정수인지 확인합니다.
 
 ```py
 isFloatStr("1") # True
 isFloatStr("123") # True
 isFloatStr("123.123") # True
 isFloatStr("HI") # False
 ```

````python
true = True
false = False

def isIntStr(str):
    try:
        b = int(str)
        if "." in str:
            return False
        return True
    except:
        return False
    
def isFloatStr(str):
    try:
        b = float(str)
        return True
    except:
        return False

def inp_int(addi):
    try:
        a = input("[입력][정수]" + addi + " ? ")
        return int(a)
    except:
        print("[에러] 정수가 들어오지 않았습니다.")
        return inp_int(addi)
    
def inp_float(addi):
    try:
        a = input("[입력][소수]" + addi + " ? ")
        return float(a)
    except:
        print("[에러] 소수가 들어오지 않았습니다.")
        return inp_float(addi)
    
def inp_char(addi):
    a = input("[입력][문자]" + addi + " ? ")
    if len(a) != 1:
        print("[에러] 한글자만 들어오지 않았습니다.")
        return inp_char(addi)
    return a

def inp_bool(name, addi):
    a = input("[입력][참거짓]" + addi + " ? ")
    if a == "0":
        return False
    if isIntStr(a):
        return True
    a = a.lower()
    if a == "t":
        return True
    if a == "f":
        return False
    if a == "true":
        return True
    if a == "false":
        return False
        

def inp(type, name = ""):
    addi = ""
    if len(name) > 0:
        addi = "[" + name + "]"
            
    if type == "int":
        return inp_int(addi)
    if type == "float":
        return inp_float(addi)
    if type == "char":
        return inp_char(addi)
    if type == "bool":
        return inp_bool(addi)
    if type == "str":
        return input("[입력][문자열]" + addi + " ? ")
    raise Exception("정의되지 않은 타입을 입력받으려고 시도했습니다. " + type)
````
